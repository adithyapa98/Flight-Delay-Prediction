```python
columns = list(['Code', 'Date', 'time', 'windspeedKmph', 'winddirDegree',
                'weatherCode', 'precipMM', 'visibility', 'pressure', 
                'cloudcover', 'DewPointF', 'WindGustKmph', 'tempF', 'WindChillF', 'humidity'])
Weather = pd.DataFrame(columns = columns)
i=0
print(columns)
```

## Forming the Weather database by parsing all files

```python
for code in os.listdir(pathw):
    
    addr = pathw + '/' + code
    #print(code)
    for month in os.listdir(addr):
        l3 = {}
        data = pd.read_json((addr + '/' + month))
        #date = 
        year, mon = month.split('-')
        if(year in list(['2016','2017'])):
            mon,_ = mon.split('.')
            if(len(mon)<2):
                mon = '0' + mon
            date = year + '-' + mon
            
            #print(date, len(data['data']['weather']))
            for day in range(len(data['data']['weather'])): 
                l3['Code'] = code
                l3['Date']=list([pd.to_datetime(date + '-' + str(day+1))])
                for time in range(24):
                    l3['time'] = list([time*100])
                    for column in columns[3:]:
                        l3[column] = list([data['data']['weather'][day]['hourly'][time][column]])
                        #print(l3)
                    df = pd.DataFrame.from_dict(l3,orient = 'columns')
                    #print(df)
                    #break
                    Weather = Weather.append(df, ignore_index = True)
                    print(i)
                    i=i+1
                    #l3 = {}
                    #print(Weather)
        else:
            continue
    
```
